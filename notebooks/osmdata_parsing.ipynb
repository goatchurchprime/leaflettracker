```python
"""
Compute the total length of highways in an osm file.
Shows how to extract the geometry of a way.
"""
import osmium as o
import sys, json

jroadsdict = { }
Dw = None
class RoadLengthHandler(o.SimpleHandler):
    def __init__(self):
        super(RoadLengthHandler, self).__init__()
        self.length = 0.0

    def way(self, w):
        global Dw
        if 'highway' in w.tags and 'name' in w.tags:
            #Dw = w
            #print(w.tags["highway"], w.tags["name"], len(w.nodes), w.nodes[0].lat, w.nodes[0].lon)
            jnodes = [ [n.lat, n.lon]  for n in w.nodes ]
            jroad = { 'highway':w.tags["highway"], "name":w.tags["name"], "nodes":jnodes }
            jroadsdict[w.id] = jroad

osmfile = "../osmdata/wavertree_map.osm"
h = RoadLengthHandler()
# As we need the geometry, the node locations need to be cached. Therefore
# set 'locations' to true.
h.apply_file(osmfile, locations=True)
fout = open("../wavertree_roads.js", "w")
fout.write("var osmroads = ")
json.dump(jroadsdict, fout, indent=2)
fout.close()
print(jroadsdict.keys())
```
